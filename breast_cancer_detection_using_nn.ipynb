```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

data collection and processing

```python
breast_cancer_dataset=sklearn.datasets.load_breast_cancer()
print(breast_cancer_dataset
)
```

```python
data_frame=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names
)
data_frame.head()
```

```python
data_frame['label']=breast_cancer_dataset.target
data_frame.tail()
```

separating the features and target

```python
x=data_frame.drop(columns='label',axis=1)
y=data_frame['label']
```

```python
print(x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_std=scaler.fit_transform(x_train)

x_test_std=scaler.fit_transform(x_test)
print(x_train_std,x_test_std)
```

building the neuRAL NETWORK

```python
import tensorflow as tft
tf.random.set_seed(3)
from tensorflow import keras
#tensorflow is a deeplearning lib developed by google.we can create nn using tensorflow.
#keras is just a wrapper of tensorflow.keras uses tf as backend
#several wgts are there. accuracy values may be changed each time . so in order to avoid that we use random_seed
```

```python
#setting up the layers of nn
#first dense is hiddenlayer, last one is op,flatten is ip layer and also to convert into sing dim layer
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(2,activation='sigmoid')
])
#no of neurans in op is equal to no of target values here we have 0,1->2
```

```python
 #compiling the nn
 model.compile(optimizer='adam',
               loss='sparse_categorical_crossentropy',
               metrics=['accuracy'])
```

one hot encoding label

```python
"""
so if u have dog the first value should be one, cat sec val shd be one and horse third val should be one
dog=[1 0 0]
cat=[0 1 0]
horse=[0 0 1]
"""
```

```python
"""
0 or 1 we use sparse categorical cross entropy
"""
```

```python
#trainig nn
history=model.fit(x_train_std,y_train,validation_split=0.1,epochs=10)
```

visualization accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','validation data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('accuracy')
plt.xlabel('loss')
plt.legend(['training data','validation data'],loc='upper right')
```

accuracy of the model on test data

```python
loss,accuracy = model.evaluate(x_test_std, y_test)
print(accuracy)
```
